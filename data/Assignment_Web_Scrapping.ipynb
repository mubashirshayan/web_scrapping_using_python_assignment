```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue?utm_source=chatgpt.com'
```

```python
url
```

```python
# params = {
#     "action": "query",
#     "format": "json",
#     "list": "search",
#     "srsearch": "Web scraping"
# }

headers = {
    "User-Agent": "Assignment_Web_Scrapping/1.0 (contact: shayanraees890@gmail.com)"
}

response = requests.get(url,headers=headers)
```

```python
soup =BeautifulSoup(response.text,"html.parser")
```

```python
print(soup.prettify())
```

```python
y = soup.find("table")
```

```python
column_header= y.find_all('th')
column_header
```

```python
column_title =[i.text.strip() for i in column_header]
```

```python
column_title
```

```python
df1 =pd.DataFrame(columns=column_title)
```

```python
df1
```

```python
rows = y.find_all('tr')
for row in rows[1:]:
    # Now 'row' is a <tr> tag. You can then find 'td' or 'th' elements within this row.
    allrow=row.find_all('td')
    rowsdata=[j.text.strip() for j in allrow]
    print(rowsdata)
    length = len(df1)
    df1.loc[length]  =rowsdata
```

```python
df1
```

```python
public_company =df1.to_csv("public_companies.csv")
```
